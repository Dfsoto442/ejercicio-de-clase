¿Que correlación existe entre cada una de las variables?
existen correlaciones negativas entre algunas variables, tales como:
1.correlación positiva entre el tiempo en pantalla y el tiempo en que el individuo pasa navegando en la plataforma Tik Tok.
2.Mayor Tiempo en pantalla se puede ver reflejado en mayores niveles de estres.
3 Mayor tiempo en pantalla y en el tiempo en tik tok menor bienestar emocional.
4 mayor cantidad de sueño mejor bienestar emocional


```python
# Define los rangos para screen_time_hours
bins = [1, 3, 6, 9, 12]
labels = ['1-3', '4-6', '7-9', '10-12']

# Crea una nueva columna con los rangos de screen_time_hours
mental_health['screen_time_range'] = pd.cut(mental_health['screen_time_hours'], bins=bins, labels=labels, right=True, include_lowest=True)

# Crea el diagrama de caja y bigotes
plt.figure(figsize=(10, 6)) # Ajusta el tamaño de la figura si es necesario
sns.boxplot(x='screen_time_range', y='stress_level', data=mental_health, order=labels)

# Añade títulos a los ejes y al gráfico
plt.xlabel('Horas de tiempo de pantalla')
plt.ylabel('Nivel de estrés')
plt.title('Nivel de estrés por rango de horas de tiempo de pantalla')

# Muestra la gráfica
plt.show()
```

con las correlaciones halladas teniendo en cuenta el nivel de estres quise indagar en como se ven estas variables reflejadas.

se determina que las personas que pasan entre 1-3 horas en pantalla presentan una disminución aproximada de un 50% en el nivel de estres comparadas con aquellas que pasan entre 10-12 horas en pantalla

```python
# Define los rangos para screen_time_hours
bins = [1, 3, 6, 9, 12]
labels = ['1-3', '4-6', '7-9', '10-12']

# Crea una nueva columna con los rangos de screen_time_hours
mental_health['screen_time_range'] = pd.cut(mental_health['screen_time_hours'], bins=bins, labels=labels, right=True, include_lowest=True)

# Crea el diagrama de caja y bigotes
plt.figure(figsize=(10, 6)) # Ajusta el tamaño de la figura si es necesario
sns.boxplot(x='screen_time_range', y='mood_score', data=mental_health, order=labels)

# Añade títulos a los ejes y al gráfico
plt.xlabel('Horas de tiempo de pantalla')
plt.ylabel('mood score')
plt.title('Nivel de bienestar emocional por rango de horas de tiempo de pantalla')

# Muestra la gráfica
plt.show()
```

Los individuos del grupo que pasan mayor cantidad de horas promedio en pantalla presentan una disminución en su bienestar emocional, respecto aquellos que pasan menos horas en promedio en pantalla. es decir, para el grupo que toma entre 1-3 Horas en pantalla presenta un bienestar emocional de 10 "mejor" y el grupo que pasa entre 10-12 horas un promedio de aproximadamente 8.5, una disminución de casi el 15% de bienestar
